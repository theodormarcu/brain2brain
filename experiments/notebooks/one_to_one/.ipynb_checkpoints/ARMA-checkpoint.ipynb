```python
train_bin_norm_paths = utils.get_file_paths("/home/tmarcu/brain2brain/brain2brain/train_676_bin_norm_2.txt")
val_bin_norm_paths = utils.get_file_paths("/home/tmarcu/brain2brain/brain2brain/val_676_bin_norm_2.txt")
# Split the train files into a training and validation set.
electrode_count = 1
target_electrode = 5
# Training Generator
train_bin_norm_generator = generators.FGenerator(file_paths=train_bin_norm_paths,
                                        lookback=50,
                                        length=10,
                                        delay=0,
                                        batch_size=100,
                                        sample_period=1,
                                        electrodes=[target_electrode],
                                        electrode_output_ix=target_electrode,
                                        shuffle=True)
# Validation Generator
val_bin_norm_generator = generators.FGenerator(file_paths=val_bin_norm_paths,
                                        lookback=50,
                                        length=10,
                                        delay=0,
                                        batch_size=5000,
                                        sample_period=1,
                                        electrodes=[target_electrode],
                                        electrode_output_ix=target_electrode,
                                        shuffle=True)
```

```python
from statsmodels.graphics.tsaplots import plot_acf


# fit model
data = train_bin_norm_generator[0][0][1]
print(data.shape)
plot_acf(data)
plt.show()
```

```python
from pandas.plotting import autocorrelation_plot
autocorrelation_plot(data)
```

```python
model = ARMA(data, order=(10, 1))
model_fit = model.fit(disp=False)
# make prediction
yhat = model_fit.predict(len(data), len(data), typ='levels')
print(model_fit.summary())
# plot residual errors
residuals = DataFrame(model_fit.resid)
residuals.plot()
plt.show()
residuals.plot(kind='kde')
plt.show()
print(residuals.describe())
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
lookback = 50
pred_length = 10
train, test = train_bin_norm_generator[0][0][0], train_bin_norm_generator[0][1][0]
test = test.reshape(pred_length)
history = list(train.reshape(lookback))
predictions = list()
for t in range(len(test)):
    model = ARIMA(history, order=(5,1,0))
    model_fit = model.fit(disp=0)
    output = model_fit.forecast()
    yhat = output[0]
    predictions.append(yhat)
    obs = test[t]
    print('predicted=%f, expected=%f' % (yhat, obs))

    history.append(obs)
predictions = np.array(predictions).reshape(pred_length)
error = mean_absolute_error(test, predictions)
print('Test MAE: %.3f' % error)
print('Test MSE: %.3f' % mean_squared_error(test, predictions))

r, p = pearsonr(test, predictions)
print(f"Correlation: {r}. P:{p}")
# plot
plt.plot(test)
plt.plot(predictions, color='red')
plt.show()
```

```python
n_rows = train_bin_norm_generator[0][0].shape[0]
lookback = 50
pred_length = 10
mae_array = np.empty(shape=(n_rows,))
mse_array = np.empty(shape=(n_rows,))
corr_array = np.empty(shape=(n_rows,))
train_batch = train_bin_norm_generator[0][0]
test_batch = train_bin_norm_generator[0][1]
for sample_ix in range(n_rows):
    print(f"Sample {sample_ix}/{n_rows}", end="\r")
    train, test = train_batch[sample_ix], test_batch[sample_ix]
    test = test.reshape(pred_length)
    history = list(train.reshape(lookback))
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=(5,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        history.append(obs)
    predictions = np.array(predictions).reshape(pred_length)
    mae = mean_absolute_error(test, predictions)
    mse = mean_squared_error(test, predictions)
    r, p = pearsonr(test, predictions)
    mae_array[sample_ix] = mae
    mse_array[sample_ix] = mse
    corr_array[sample_ix] = r
print(f"Batch MAE: {np.mean(mae_array)}")
print(f"Batch MSE: {np.mean(mse_array)}")
z_mean = np.mean(corr_array)
batch_r = np.tanh(z_mean)
print(f"Batch Correlation: {batch_r}")
```

```python
k = 5
top_k = mae_array.argsort()[:k]
for i in top_k:
    train, test = train_batch[i], test_batch[i]
    test = test.reshape(pred_length)
    history = list(train.reshape(lookback))
    predictions = list()
    for t in range(len(test)):
        model = ARIMA(history, order=(5,1,0))
        model_fit = model.fit(disp=0)
        output = model_fit.forecast()
        yhat = output[0]
        predictions.append(yhat)
        obs = test[t]
        print('predicted=%f, expected=%f' % (yhat, obs))
        history.append(obs)
    predictions = np.array(predictions).reshape(pred_length)
    mae = mean_absolute_error(test, predictions)
    mse = mean_squared_error(test, predictions)
    print('Test MAE: %.3f' % mae)
    print('Test MSE: %.3f' % mse)
    r, p = pearsonr(test, predictions)
    print(f"Corr: {r}")
    plt.plot(test)
    plt.plot(predictions, color='red')
    plt.show()
```
